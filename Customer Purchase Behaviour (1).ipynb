```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Discount Applied', y='Purchase Amount (USD)', data=df, palette='coolwarm')

plt.xlabel("Discount (Yes/No)")
plt.ylabel("Sales")
plt.title("Comparison of Sales with and without Discount")

plt.show()
```

```python
from scipy.stats import ttest_ind

sales_yes = df[df['Discount Applied'] == 'Yes']['Purchase Amount (USD)']
sales_no = df[df['Discount Applied'] == 'No']['Purchase Amount (USD)']

t_stat, p_value = ttest_ind(sales_yes, sales_no, equal_var=False)  # Welch’s t-test (unequal sample sizes)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The difference in sales is statistically significant. Discounts affect sales!")
else:
    print("No significant difference in sales. Discounts may not have a strong impact.")
```

```python
sales_by_subscription = df.groupby('Subscription Status')['Purchase Amount (USD)'].agg(['sum', 'mean', 'count'])
print(sales_by_subscription)
```

```python
subscription_yes = df[df['Subscription Status'] == 'Yes']['Purchase Amount (USD)']
subscription_no = df[df['Subscription Status'] == 'No']['Purchase Amount (USD)']

t_stat, p_value = ttest_ind(subscription_yes, subscription_no, equal_var=False)  # Welch’s t-test (unequal sample sizes)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The difference in sales is statistically significant. Subscription Status affect sales!")
else:
    print("No significant difference in sales. Subscription Status may not have a strong impact.")
```
